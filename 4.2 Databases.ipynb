# Основні поняття БД
**Constraints** - обмеження  
Обмеження дозволяють додатково обмежити область атрибута. Наприклад, обмеження може обмежувати заданий цілий атрибут значеннями від 1 до 10. Обмеження забезпечують один метод реалізації бізнес-правил у базі даних та підтримують подальше використання даних у рівні програми.  
**Primary key** - первинний ключ  
Первинний ключ однозначно вказує запис всередині таблиці. Для того, щоб атрибут був хорошим первинним ключем, він не повинен повторюватися. Хоча природні атрибути (атрибути, що використовуються для опису даних, що вводяться) іноді є хорошими первинними ключами, замість них часто використовуються сурогатні ключі. Сурогатний ключ (surrogate key) - це штучний атрибут, присвоєний об'єкту, який однозначно його ідентифікує  
**Foreign key** - зовнішній ключ  
Зовнішній ключ - це поле у реляційній таблиці, яке відповідає стовпцю первинного ключа іншої таблиці. Він пов'язує дві таблиці. Зовнішні ключі не повинні мати унікальних значень у відношенні.  
**Stored procedures** - збережені процедури  
Збережена процедура - це виконуваний код, який асоціюється з базою даних і, як правило, зберігається. Збережені процедури не є частиною моделі реляційних баз даних, але всі комерційні реалізації включають їх.  
**Index** - індекс  
Індекс — об’єкт бази даних, що створений з метою підвищення ефективності
виконання запитів.



# SQL

Майже кожна реляційна СУБД підтримує мову запитів під назвою SQL  
**SQL** (Structured query language — мова структурованих запитів) — декларативна мова програмування для взаємодії користувача з базами даних, що застосовується для формування запитів, оновлення і керування реляційними БД створення схеми бази даних та її модифікації, системи контролю за доступом до бази даних.  
Дана мова складається з таких компонентів:  
•	DQL (Data Query Language) - створення запитів  
•	DDL (Data Definition Language) — робота зі структурою бази.  
•	DML (Data Manipulation Language) — робота з рядками.  
•	DCL (Data Control Language) — робота з правами.  
•	TCL (Transaction Control Language) — робота з транзакціями.  

## Data Definition Language//Мова визначення даних

**CREATE TABLE**	Створити нову таблицю в БД	
**DROP TABLE**	    Видалити таблицю з БД	
**ALTER TABLE**	    Змінити структуру існуючої БД	
**CREATE VIEW**	    Створити віртуальну таблицю, яка відповідає певному sql-запиту	
**ALTER VIEW**	    Змінити існуючий розріз (view) даних	
**DROP VIEW**	    Видалити розріз (view)	
**CREATE INDEX**	Створити індекс для таблиці	
**DROP INDEX**	    Видалити існуючий індекс	

## Data Manipulation Language//Мова маніпулювання даними

**DELETE** Видалити один або декілька рядків за заданим фільтром із існуючої таблиці. Даний оператор  узгоджується із правилом цілісності даних, тому не завжди може бути виконаний  
**INSERT** Вставити один рядок в таблицю. В різних модифікаціях є можливість вставляти масив даних  
**UPDATE** Оновити рядок із таблиці за задними фільтрами  

## Transaction Control Language//Мова керування транзакціями

**COMMIT**		Завершує транзакцію, тобто послідовність дій чи обробки/запису інформації  
**ROLLBACK**		Відмінити зміни, що виконанні в ході виконання транзакції  
**SAVEPOINT**		Зберегти проміжний стан бази даних, щоб при потребі можна було б повернутися в цю точку  

## Data Query Language//Мова запиту даних

**SELECT**		Перелічує поля, які містять потрібні дані  
**FROM**		Перелічує таблиці, у яких містяться поля, зазначені в реченні `SELECT`  
**WHERE**		Визначає умови для поля, яким мають відповідати всі записи, щоб бути включеними до результатів  
**ORDER BY**		Визначає спосіб сортування результатів  
**GROUPBY**		В інструкції `SQL`, яка містить агрегатні функції, перелічує поля, не зведені в реченні `SELECT`  

### Функції та оператори SQL

**SUM**			Повертає суму значень в стовбці  
**MIN/МАХ**			Повертає максмаильне(мінімальне) значення в стовбці  
**AVG**			Повертає середнє значення  
**COUNT** Повертає кількість записів в стовбці  
**DISTINCT** Оператор для унікальних значень   
**IN**			Для перебору значень (використовується в `WHERE`)  <br>
**BETWEEN**			Для виведення даних із певного діапазону (використовується в `WHERE`) <br> 
**LIKE**			Створює шаблон пошуку даних (використовується в `WHERE`)   

### JOIN SQL

Join — операція з'єднання таблиць в SQL, яка сполучає дві таблиці в реляційній базі даних, утворюючи нову тимчасову таблицю, яку інколи називають «з'єднаною таблицею».

Згідно з ANSI-стандартом, в SQL існують такі типи з'єднання: 
* внутрішнє — INNER, 
* зовнішнє — OUTER та 
* перехресне — CROSS. 

Зовнішнє з’єднання поділяється на ліве — LEFT OUTER, праве — RIGHT OUTER та повне — FULL OUTER. 

Особливим випадком є з’єднання таблиці з собою, що має назву самоз'єднання
З математичної точки зору операція з’єднання є фундаментальною операцією реляційної алгебри.
