Розглянемо **послідовність**, задану формулою:
$$\lim_{x \to \infty } \frac{5n^3-3n^2}{n^3+1}$$
$n$ прагне нескінченності, це означає, що $n$ приймає всілякі натуральні значення $\mathbb N$.

Вираз, записаний вище, означає число, яке вийшло б якби ми «підставили» нескінченність у наш дріб. Тобто. значення дробу в абстрактній нескінченності не дуже сильно відрізнятиметься, якщо підставити туди мільйон. **Не можна підібрати** такого натурального $n$, щоб вираз дорівнював своїй границі (тому **прагне, але, на жаль, не досягає**).

Спробуймо «дістатись до нескінченності» за допомогою обчислень. Отже код на Python:

```python
def seq(n):
    return (5 * n ** 3 - 3 * n**2) / (n**3 + 1)
for n in range(1, 1000 + 1):
    print(n, seq(n))
```

Вже на 1000-му елементі значення послідовності так близько підібралося до 5.
Повторимо вже в проміжку від мільйона

```python
for n in range(1000000, 1000010 + 1):
    print(n, seq(n))
```

Як бачимо нічого не змінюється, значення послідовності наближається 5

Перші завдання, з якими можна зіткнутися — це доказ того, що **якесь число** є **границею** послідовності визначення. Завдання неприємне тим, що дійсно перевіряє, наскільки ви відчули визначення і постійно викликає складності. Здавалося б класика: довести, що границя$\frac{1}{n} $ дорівнює 0 за визначенням.

```python
import math
epsilon = float(input(f'epsilon :'))
n = 1
limit = 0 # 
def sequence(n):
    return 1 / n
while abs(sequence(n) - limit) > epsilon:
    print(n, sequence(n))
    n += 1
print("N(ε) найден: ", n)
```

# Список використаних позначень

<img src='https://disted.edu.vn.ua/media/images/porkhun/10_inf/chislovi_mnozini/001.png'>

Позначення $\mathbb N$, $\mathbb Z$, $\mathbb Q$, $\mathbb R$, $[a,b][a,b]$, $(a,b)(a,b)$, $[a,b)[a,b)$ , $(a,b](a,b]$ і $\sum\limits_{k=m}^n a_k$
передбачаються відомими заздалегідь.


* $\mathbb N$ - множина натуральних чисел: 1, 2, 3, ....
* $\mathbb Z$ - множина цілих чисел, окрім натуральних чисел до неї входять від'ємні та нуль
* $\mathbb Q$ - множина раціональних чисел, тобто дробів виду $\frac{m}{n}$, де $m\in\mathbb Z$ і $n\in\mathbb N$.
* $\mathbb R$ - множина дійсних чисел.
* $\mathbb I$ Iррацiональнi числа — числа, що не є рацiональними, тобто не можуть бути вираженi вiдношенням цiлих чисел. Мають вигляд нескiнченних десяткових дробiв. Всi знають число $\pi$, яке i є iррацiональним 3,14159265... i так далi.
<br>
* Запис $\{ x\in X : \text{ деяка умова на } x \}$ означає підмножину множини X, що складається з елементів, що задовольняють зазначеній умові.
<br>
<br>
* Записи $A\cup B$ і $A\cap B$ позначають відповідно об'єднання та перетин множин A і B.
<br>
* Записи $\bigcup\limits_{k=1}^n A_k$  і $\bigcap\limits_{k=1}^n A_k$ позначають відповідно об'єднання та перетин всіх множин $A_1,A_2, \dots,A_n.$
<br>
<br>
* $[a,b]$ — відрізок між числами a та b, тобто множина дійсних чисел, що лежать між числами a та b, включаючи самі числа a та b $([a,b] = \{ x\in \mathbb R : a\leqslant x \leqslant b \})$ При a = b відрізок складається з однієї точки і називається виродженим відрізком.
<br>
<br>
* $(a,b)$ — інтервал між числами a і b, тобто множина дійсних чисел, що лежать строго між a і b ($(a,b) = \{ x\in \mathbb R : a \lt x < b$)
<br>
* $[a,b)$ і $(a,b]$ — напівінтервали: $[a,b) = \{ x\in \mathbb R : a\leqslant x \lt b \}$ і $(a,b] = \{ x\in \mathbb R : a \lt x\leqslant b \}$
<br>
<br>
* $[a,+\infty)$ — промінь: $[a,+\infty) = \{ x\in \mathbb R : a\leqslant x \}$
* $(a,+\infty)$ — промінь: $(a,+\infty) = \{ x\in \mathbb R : a \lt x \}$
* $(-\infty,b]$ — промінь: $(-\infty,b] = \{ x\in \mathbb R : x\leqslant b \}$
* $(-\infty,b)$ — промінь: $(-\infty,b) = \{ x\in \mathbb R : x \lt b \}$
<br>
<br>
* Відрізок, інтервал та напівінтервали називаються проміжками. Проміжок будь-якого з чотирьох типів позначається $\langle a,b\rangle$. У межах одного твердження запис $\langle a,b\rangle$ завжди позначає той самий проміжок.
<br>
* $\langle a, b)$ - будь-який з двох проміжків $(a, b)$ і $[a, b)$; $\langle a,b]$ - будь-який з двох проміжків $(a,b]$ і $[a,b]$ і т.д.
<br>
<br>
* $\forall$ - квантор загальності, використовується для скороченого запису замість слів "кожен", "будь-який", або "для всякого", "для будь-якого" і т.п.
<br>
<br>
* $\exists$ - квантор існування, використовується для скороченого запису замість слів "знайдеться", "існує" і т.п.
<br>
<br>
* $\sum\limits_{k=m}^n a_k$ - Сума чисел $a_{k}$ по $k$ від $m$ до $n$, тобто $a_m+a_{m+1}+a_{m+2}+\ldots + a_n$
<br>
<br>
* $f: X \to Y$ — функція, задана на множині $X$, множині значень якої лежить у $Y$ (але необов'язково з ним збігається).
